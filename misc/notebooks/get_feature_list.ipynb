```python
def download_premade_features(name):
    features_path = os.path.join("pregenerated-features", name, "features")
    features, labels = download_object(features_path)
    return features, labels


def download_object(path):
    bucket_path = f"gs://{path}"
    with gfile.GFile(bucket_path, "rb") as wf:
        obj = joblib.load(wf)
    return obj
```

```python
df, labels = download_premade_features("emip-fixations")
```

```python
def filter_df(df, regexes):
    combined_regex = "|".join(regexes)
    regex = f"^(?:{combined_regex})"
    df = df.loc[:, df.columns.str.contains(regex)]
    return df
```

```python
feature_regexes = [
    "^duration_rolling__arma__*",
    "saccade_duration_rolling__lhipa",
]
```
