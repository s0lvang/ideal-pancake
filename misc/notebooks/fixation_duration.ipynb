```python
def get_saccade_length(data, x, y):
    coordinates = data.loc[:, [x, y]].to_numpy()
    shifted_coordinates = np.roll(coordinates, -1, axis=0).tolist()
    shifted_coordinates[-1] = coordinates[-1]
    saccades = pd.Series([np.linalg.norm(a - b) for a, b in zip(coordinates, shifted_coordinates)])
    return saccades[saccades > 0.001]
```
